```python
resources = {
    'water': 300,
'milk': 100,
'coffee': 76,
'money': 2.5

}

menu = {
    "espresso": {
        "ingrediant":{
            "water":50,
            "coffee":18
        },
        "cost":1.5
    },

    "latte": {
        "ingrediant":{
            "water":200,
            "coffee" : 24,
            "milk":150
        },
        "cost":2.5
    },

    "capaccuino": {
        "ingrediant":{
            "water":250,
            "coffee" : 24,
            "milk":100
        },
        "cost":3.0
    },

}

QUARTER = 0.25
DIMES = 0.10
NICKLES = 0.05
PENNIES = 0.01

print(menu["capaccuino"]["ingrediant"]["water"])

def check_resource_available(drink):
    for key in menu[drink]["ingrediant"]:
        if(menu[drink]["ingrediant"][key]<= resources[key]):
            return True
        else :
            print(f"Sorry, there is not enough {key}") 
            return False

def make_drink(drink):
    for key in menu[drink]["ingrediant"]:
        rest =  resources[key] - menu[drink]["ingrediant"][key] 
        resources[key] = rest

def calculate_amount(qrts, dime, nik, penn, drink):
    total =(qrts*QUARTER)+(dime*DIMES)+(nik*NICKLES)+(penn*PENNIES)
    drink_cost = menu[drink]["cost"]
    if(total<drink_cost):
        print(f"Sorry, that's not enough money. Money refunded")
        return False
    elif(total == drink_cost):
        print(f"Amount is sufficient")
        resources["cost"]+= total
        return True
    else:
        change = round(total - drink_cost, 2)
        resources["cost"]+= total
        print(f"Here is the change ${change} in change")
        return True



on = True

while on:
    choice = input("what would you like to drink(capccuino/espresso/latte)").lower()
    if(choice=='off'):
        on = False
    elif(choice == 'report'):
        print(resources)
    else:
        resource_avail = check_resource_available(choice)
        if not resource_avail:
            on = False
        else:
            print("Please insert coins")
            qrts = int(input("How many quarters"))
            dimes = int(input("How many dimes"))
            nickles = int(input("How many nickles"))
            penny = int(input("How many pennies"))
            amnt = calculate_amount(qrts, dimes,nickles, penny, choice)
            if(not amnt):
                on = False
            else:
                make_drink(choice)
                print(f"Here is your {choice}, Enjoy!")
    
```
